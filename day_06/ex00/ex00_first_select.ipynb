```python
import sqlite3
import pandas as pd

try:
    sqlite_connection = sqlite3.connect('data/checking-logs.sqlite.sqlite')
except sqlite3.Error as error:
    print("Ошибка при подключении к sqlite", error)
```

```python
query = 'PRAGMA table_info(pageviews);'
pd.io.sql.read_sql(query, sqlite_connection)
```

```python
query = 'SELECT * from pageviews;'
pd.io.sql.read_sql(query, sqlite_connection).head(10)
```

```python
query = "SELECT uid,datetime from pageviews WHERE uid LIKE 'user_%' ORDER BY uid ASC;"
res = pd.io.sql.read_sql(query, sqlite_connection, index_col='datetime', parse_dates=['datetime'])
res.name = 'pageviews'
res.head(10)
```

```python
sqlite_connection.close()
```
