```python
import sqlite3
import pandas as pd

try:
    sqlite_connection = sqlite3.connect('data/checking-logs.sqlite.sqlite')
except sqlite3.Error as error:
    print("Ошибка при подключении к sqlite", error)
print('УСПЕШНО')
```

```python
request = \
"""
SELECT checker.uid, labname, checker.timestamp AS first_commit_ts, pageviews.datetime AS first_view_ts FROM checker
LEFT JOIN pageviews ON checker.uid == pageviews.uid
WHERE status=='ready' AND numTrials==1 
AND labname in ('laba04', 'laba04s', 'laba05', 'laba06','laba06s', 'project1')
AND checker.uid LIKE 'user_%'
AND (pageviews.datetime = (SELECT MIN(pageviews.datetime)
                           FROM pageviews
                           WHERE uid=checker.uid)
     OR pageviews.datetime IS NULL)
;
"""
res = pd.io.sql.read_sql(request, sqlite_connection, parse_dates=['first_commit_ts', 'first_view_ts'])
res
```

```python
res.info()
```

```python
# test = res.where(res.first_view_ts.notnull() == True).dropna()[['uid','first_view_ts']]
test = res.where(res.first_view_ts.notnull() == True).dropna()
```

```python
test
```

```python
# control = res.where(res.first_view_ts.notnull() == False).dropna(subset='uid')[['uid','first_view_ts']]
control = res.where(res.first_view_ts.notnull() == False).dropna(subset=['uid'])
```

```python
control
```

```python
control = control.fillna(test.first_view_ts.dropna().mean())
```

```python
test.to_sql('test', sqlite_connection)
control.to_sql('control', sqlite_connection)
```

```python
sqlite_connection.close()
```
